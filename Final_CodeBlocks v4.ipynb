<font color="red">As per NTCIR18-Lifelog6 task,</font>
The lifelog task appears to be a benchmark for evaluating how well systems can process and understand personal data collected through lifelogging applications. Here's a breakdown of the three subtasks according to NTCIR:
<br>
**1. Lifelog Semantic Access (LSAT):**
Focuses on retrieving specific moments from a user's lifelog based on a query.
The query can be in text format (written description), visual (image), or a combination of both.
This subtask is similar to a traditional search engine but specifically targets personal data.
<br>
**2. Lifelog Insight (LIT):**
Aims to uncover hidden patterns and insights from a user's lifelog data.
This year's focus is on automatically detecting everyday activities (like sleeping, eating, exercising) from a user's 18-month lifelog data.
The goal is to develop systems that can analyze large amounts of lifelog data and automatically identify recurring patterns in a user's daily routines.
<br>
**3. Lifelog Question Answering (LQAT):**
Evaluates how well systems can answer questions directly based on a user's lifelog information.
Participants are provided with a dataset containing questions and corresponding answers related to a user's 85-day lifelog data.
The goal is to develop systems that can understand and answer questions about a user's experiences and activities as reflected in their lifelog data.
<br>
In essence, these subtasks explore different aspects of lifelog data analysis:<br>
**LSAT - **Targeted retrieval of specific moments.<br>
**LIT - **Uncovering broader patterns and insights.<br>
**LQAT -** Answering questions directly based on the data.<br>
By tackling these subtasks, researchers can evaluate and improve the capabilities of systems designed to understand and utilize personal lifelog data.

```python
MARGIN = 10  # pixels
ROW_SIZE = 10  # pixels
FONT_SIZE = 1
FONT_THICKNESS = 1
TEXT_COLOR = (255, 0, 0)  # red

import numpy as np
import cv2
import numpy as np
from collections import Counter
# STEP 1: Import the necessary modules.
import numpy as np
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
import cv2
from google.colab.patches import cv2_imshow
# import dependencies
from IPython.display import display, Javascript, Image
from google.colab.output import eval_js
from base64 import b64decode, b64encode
import cv2
import numpy as np
import PIL
import io
import html
import time
import csv
import pandas as pd
import os

def FinalPrediction(listForExcel):
  # Flatten the list of lists
  flattened_x = [item for sublist in listForExcel for item in sublist]
  # Count the occurrences of each string
  counter = Counter(flattened_x)
  # Find the most common string
  most_common = counter.most_common(1)
  # Output the result
  if most_common:
      return most_common[0][0]
  else:
      return "NA"

def visualize(image,detection_result) -> np.ndarray:
  """Draws bounding boxes on the input image and return it.
  Args:
    image: The input RGB image.
    detection_result: The list of all "Detection" entities to be visualize.
  Returns:
    Image with bounding boxes.
  """
  for detection in detection_result.detections:
    # Draw bounding_box
    bbox = detection.bounding_box
    start_point = bbox.origin_x, bbox.origin_y
    end_point = bbox.origin_x + bbox.width, bbox.origin_y + bbox.height
    cv2.rectangle(image, start_point, end_point, TEXT_COLOR, 3)

    # Draw label and score
    category = detection.categories[0]
    category_name = category.category_name
    probability = round(category.score, 2)
    result_text = category_name + ' (' + str(probability) + ')'
    text_location = (MARGIN + bbox.origin_x,
                     MARGIN + ROW_SIZE + bbox.origin_y)
    cv2.putText(image, result_text, text_location, cv2.FONT_HERSHEY_PLAIN,
                FONT_SIZE, TEXT_COLOR, FONT_THICKNESS)
  return image

def ImageClassification(gray):
  # STEP 2: Create an ObjectDetector object.
  base_options = python.BaseOptions(model_asset_path='efficientdet.tflite')
  options = vision.ObjectDetectorOptions(base_options=base_options,
                                        score_threshold=0.5)
  detector = vision.ObjectDetector.create_from_options(options)
  # STEP 3: Load the input image.
  #print(gray)
  #print(type(gray))
  #image = mp.Image.create_from_file(gray)
  image = mp.Image(
      image_format=mp.ImageFormat.SRGB,  # Assuming your image is in SRGB format
      data=gray
  )

  # STEP 4: Detect objects in the input image.
  detection_result = detector.detect(image)
  # STEP 5: Process the detection result. In this case, visualize it.
  image_copy = np.copy(image.numpy_view())
  annotated_image = visualize(image_copy, detection_result)
  rgb_annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
  cv2_imshow(rgb_annotated_image)
  y = detector.detect(image)
  #print(type(y))
  #print(y)
  x = str(y)
  #print(x)
  import re
  # Define a regular expression pattern to extract category names
  pattern = r"category_name='(\w+)'"
  # Find all matches of the pattern in the string x
  matches = re.findall(pattern, x)
  # Print the extracted category names
  return matches

# function to convert the JavaScript object into an OpenCV image
def js_to_image(js_reply):
  """
  Params: js_reply: JavaScript object containing image from webcam
  Returns: img: OpenCV BGR image
  """
  # decode base64 image
  image_bytes = b64decode(js_reply.split(',')[1])
  # convert bytes to numpy array
  jpg_as_np = np.frombuffer(image_bytes, dtype=np.uint8)
  # decode numpy array into OpenCV BGR image
  img = cv2.imdecode(jpg_as_np, flags=1)

  return img

# function to convert OpenCV Rectangle bounding box image into base64 byte string to be overlayed on video stream
def bbox_to_bytes(bbox_array):
  """
  Params: box_array: Numpy array (pixels) containing rectangle to overlay on video stream.
  Returns: bytes: Base64 image byte string
  """
  # convert array into PIL image
  bbox_PIL = PIL.Image.fromarray(bbox_array, 'RGBA')
  iobuf = io.BytesIO()
  # format bbox into png for return
  bbox_PIL.save(iobuf, format='png')
  # format return string
  bbox_bytes = 'data:image/png;base64,{}'.format((str(b64encode(iobuf.getvalue()), 'utf-8')))

  return bbox_bytes


# initialize the Haar Cascade face detection model
face_cascade = cv2.CascadeClassifier(cv2.samples.findFile(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'))


# JavaScript to properly create our live video stream using our webcam as input
def video_stream():
  js = Javascript('''
    var video;
    var div = null;
    var stream;
    var captureCanvas;
    var imgElement;
    var labelElement;

    var pendingResolve = null;
    var shutdown = false;

    function removeDom() {
       stream.getVideoTracks()[0].stop();
       video.remove();
       div.remove();
       video = null;
       div = null;
       stream = null;
       imgElement = null;
       captureCanvas = null;
       labelElement = null;
    }

    function onAnimationFrame() {
      if (!shutdown) {
        window.requestAnimationFrame(onAnimationFrame);
      }
      if (pendingResolve) {
        var result = "";
        if (!shutdown) {
          captureCanvas.getContext('2d').drawImage(video, 0, 0, 640, 480);
          result = captureCanvas.toDataURL('image/jpeg', 0.8)
        }
        var lp = pendingResolve;
        pendingResolve = null;
        lp(result);
      }
    }

    async function createDom() {
      if (div !== null) {
        return stream;
      }

      div = document.createElement('div');
      div.style.border = '2px solid black';
      div.style.padding = '3px';
      div.style.width = '100%';
      div.style.maxWidth = '600px';
      document.body.appendChild(div);

      const modelOut = document.createElement('div');
      modelOut.innerHTML = "<span>Status:</span>";
      labelElement = document.createElement('span');
      labelElement.innerText = 'No data';
      labelElement.style.fontWeight = 'bold';
      modelOut.appendChild(labelElement);
      div.appendChild(modelOut);

      video = document.createElement('video');
      video.style.display = 'block';
      video.width = div.clientWidth - 6;
      video.setAttribute('playsinline', '');
      video.onclick = () => { shutdown = true; };
      stream = await navigator.mediaDevices.getUserMedia(
          {video: { facingMode: "environment"}});
      div.appendChild(video);

      imgElement = document.createElement('img');
      imgElement.style.position = 'absolute';
      imgElement.style.zIndex = 1;
      imgElement.onclick = () => { shutdown = true; };
      div.appendChild(imgElement);

      const instruction = document.createElement('div');
      instruction.innerHTML =
          '<span style="color: red; font-weight: bold;">' +
          'When finished, click here or on the video to stop this demo</span>';
      div.appendChild(instruction);
      instruction.onclick = () => { shutdown = true; };
      document.body.onclick = () => { shutdown = true; };


      video.srcObject = stream;
      await video.play();

      captureCanvas = document.createElement('canvas');
      //captureCanvas.width = 640; //video.videoWidth;
      //captureCanvas.height = 480; //video.videoHeight;
      captureCanvas.width = 640; //video.videoWidth;
      captureCanvas.height = 480; //video.videoHeight;
      window.requestAnimationFrame(onAnimationFrame);

      return stream;
    }
    async function stream_frame(label, imgData) {
      if (shutdown) {
        removeDom();
        shutdown = false;
        return '';
      }

      var preCreate = Date.now();
      stream = await createDom();

      var preShow = Date.now();
      if (label != "") {
        labelElement.innerHTML = label;
      }

      if (imgData != "") {
        var videoRect = video.getClientRects()[0];
        imgElement.style.top = videoRect.top + "px";
        imgElement.style.left = videoRect.left + "px";
        imgElement.style.width = videoRect.width + "px";
        imgElement.style.height = videoRect.height + "px";
        imgElement.src = imgData;
      }

      var preCapture = Date.now();
      var result = await new Promise(function(resolve, reject) {
        pendingResolve = resolve;
      });
      shutdown = false;

      return {'create': preShow - preCreate,
              'show': preCapture - preShow,
              'capture': Date.now() - preCapture,
              'img': result};
    }
    ''')

  display(js)


def video_frame(label, bbox):
  data = eval_js('stream_frame("{}", "{}")'.format(label, bbox))
  #print("data--------",data)
  return data





def write_to_excel(file_path, data):
    if isinstance(data, pd.DataFrame):
        if os.path.exists(file_path):
            # If the file already exists, append data to it
            existing_data = pd.read_excel(file_path)
            updated_data = pd.concat([existing_data, data], ignore_index=True)
            updated_data.to_excel(file_path, index=False)
            print("Data appended to existing Excel file:", file_path)
        else:
            # If the file doesn't exist, create a new one
            data.to_excel(file_path, index=False)
            print("Data written to new Excel file:", file_path)
    elif isinstance(data, list):
        with open(file_path, 'a', newline='') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(data)
            print("List appended to existing CSV file:", file_path)
    else:
        print("Unsupported data format. Expected DataFrame or list.")

def measure_brightness(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Calculate the average brightness using NumPy
    avg_brightness = np.mean(gray_image)
    return avg_brightness

def assess_brightness(brightness):
    if brightness > 100:  # You can adjust this threshold as needed
        return "High"
    elif brightness > 50:
        return "Moderate"
    else:
        return "Low"


all_data = []
all_brightness_data = []
file_path = '/content/drive/My Drive/predictions.xlsx'

try:
  from google.colab import drive
  drive.mount('/content/drive')
  # start streaming video from webcam
  video_stream()
  # label for video
  label_html = 'Capturing...'
  # initialze bounding box to empty
  bbox = ''
  count = 0
  while True:
    i=0
    listForExcel = []
    brightness_data = []
    while i<5:
      js_reply = video_frame(label_html, bbox)
      if not js_reply:
          break

      # convert JS response to OpenCV Image
      img = js_to_image(js_reply["img"])

      # create transparent overlay for bounding box
      bbox_array = np.zeros([480,640,4], dtype=np.uint8)

      # grayscale image for face detection
      gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
      #print(gray)
      matches = ImageClassification(img)
      print(matches)
      print("Value of i = ", i)
      listForExcel.append(matches)
      brightness_data.append(measure_brightness(img))

      # get face region coordinates
      faces = face_cascade.detectMultiScale(gray)
      # get face bounding box for overlay
      for (x,y,w,h) in faces:
        bbox_array = cv2.rectangle(bbox_array,(x,y),(x+w,y+h),(255,0,0),2)

      bbox_array[:,:,3] = (bbox_array.max(axis = 2) > 0 ).astype(int) * 255
      # convert overlay of bbox into bytes
      bbox_bytes = bbox_to_bytes(bbox_array)
      # update bbox so next frame gets new overlay
      bbox = bbox_bytes
      i+=1
    print("==========================STORED IN EXCEL==========================", listForExcel)
    print("==========================STORED IN EXCEL==========================", FinalPrediction(listForExcel))
    # Store the result in Excel
    prediction = FinalPrediction(listForExcel)
    timestamp = pd.Timestamp.now() + pd.Timedelta(hours=5, minutes=30)

    print("brightness_data :: ", brightness_data)
    #all_brightness_data.append()

    data = {'Timestamp': timestamp, 'Prediction': prediction, 'Brightness': sum(brightness_data)/len(brightness_data)}
    all_data.append(data)


except Exception as e:
  print("Video Capturing Stopped.", e)
  # Convert the list of dictionaries to a DataFrame
  print("=====================Content of all_data=======================", all_data)
  all_data_df = pd.DataFrame(all_data)
  print("=====================Content of all_data_df=======================\n", all_data_df)
  # Write the DataFrame to the Excel file
  write_to_excel(file_path, all_data_df)
```

```python
def stepCountReportGenerationInteractive():
  import pandas as pd
  import plotly.graph_objects as go
  from datetime import datetime

  # Read data from the Excel file
  excel_file = step_count_data_file_path
  df = pd.read_excel(excel_file)

  # Define thresholds and corresponding colors
  thresholds = [5000, 7500, 10000, 12500]
  colors = ['#E74C3C', '#F5B7B1', '#F5B7B1', '#2ECC71']

  # Extract values and end times from the DataFrame
  values = df['Value']
  end_times = pd.to_datetime(df['End Date'])

  # Calculate average value
  average_value = values.mean()

  # Create traces for step count data
  data = []
  for i, value in enumerate(values):
      # Determine color based on step count range
      color_index = sum(value >= threshold for threshold in thresholds)
      color = colors[color_index]

      trace = go.Scatter(
          x=[end_times[i]],
          y=[value],
          mode='markers',
          marker=dict(color=color),
          name=f'Step Count: {value}',
          text=f'End Time: {end_times[i]}<br>Step Count: {value}',
          hoverinfo='text'
      )
      data.append(trace)

  # Create traces for average line, highest value, and lowest value
  highest_value = max(values)
  lowest_value = min(values)

  average_line = go.Scatter(
      x=end_times,
      y=[average_value] * len(end_times),
      mode='lines',
      line=dict(color='blue', dash='dash'),
      name=f'Average Step Count: {average_value}',
      hoverinfo='skip'
  )

  highest_point = go.Scatter(
      x=[end_times[values.idxmax()]],
      y=[highest_value],
      mode='markers',
      marker=dict(color='red', size=10),
      name=f'Highest Step Count: {highest_value}',
      text=f'End Time: {end_times[values.idxmax()]}<br>Highest Step Count: {highest_value}',
      hoverinfo='text'
  )

  lowest_point = go.Scatter(
      x=[end_times[values.idxmin()]],
      y=[lowest_value],
      mode='markers',
      marker=dict(color='green', size=10),
      name=f'Lowest Step Count: {lowest_value}',
      text=f'End Time: {end_times[values.idxmin()]}<br>Lowest Step Count: {lowest_value}',
      hoverinfo='text'
  )

  # Create layout
  layout = go.Layout(
      title='Step Count Over Time',
      xaxis=dict(title='End Time'),
      yaxis=dict(title='Step Count'),
      hovermode='closest',
      legend=dict(x=0.5, y=-0.2, orientation='h'),
      plot_bgcolor='rgba(0, 0, 0, 0)',
      paper_bgcolor='rgba(0, 0, 0, 1)',
      font=dict(color='white')
  )

  # Plot the graph
  fig = go.Figure(data=data + [average_line, highest_point, lowest_point], layout=layout)
  fig.show()

stepCountReportGenerationInteractive()
```

```python
import pandas as pd
import requests

def get_heart_minutes_count():
    url = "https://v1.nocodeapi.com/satyakidas/fit/eSGdIxlNTKrflNTk/aggregatesDatasets"
    params = {
        "dataTypeName": "heart_minutes",  # Only request heart_minutes data
        "timePeriod": "30days"
    }
    headers = {
        "Content-Type": "application/json"
    }

    try:
        response = requests.get(url, headers=headers, params=params)
        response.raise_for_status()  # Raise an exception for 4xx or 5xx status codes
        data = response.json()

        # Extract and return only the heart_minutes value
        if 'heart_minutes' in data:
            return data['heart_minutes']
        else:
            print("Error: 'heart_minutes' not found in response data")
            return None

    except requests.exceptions.RequestException as e:
        print("Error:", e)
        return None

# Call the function to retrieve heart minutes
heart_minutes = get_heart_minutes_count()

if heart_minutes is not None:
    # Extracting values and end times
    values = [entry['value'] for entry in heart_minutes]
    end_times = [entry['endTime'] for entry in heart_minutes]

    # Create a DataFrame to store the new data
    new_data = pd.DataFrame({'Value': values, 'New End Date': end_times})

    # Read the existing data from the Excel file if it exists
    excel_file = 'heart_minutes_data.xlsx'
    try:
        existing_data = pd.read_excel(excel_file)
        # Filter out data that already exists in the Excel file
        new_data = new_data[~new_data['New End Date'].isin(existing_data['New End Date'])]
        # Append the new data to the existing DataFrame
        updated_data = pd.concat([existing_data, new_data], ignore_index=True)
    except FileNotFoundError:
        # If the file doesn't exist, use only the new data
        updated_data = new_data

    # Save the updated DataFrame to the Excel file
    updated_data.to_excel(excel_file, index=False)

    print("Heart minutes data saved to:", excel_file)
else:
    print("No data retrieved.")
```

```python
def TextMoodPrediction(INPUT_TEXT):
  from mediapipe.tasks import python
  from mediapipe.tasks.python import text
  try:
    # STEP 2: Create an TextClassifier object.
    base_options = python.BaseOptions(model_asset_path="classifier.tflite")
    options = text.TextClassifierOptions(base_options=base_options)
    classifier = text.TextClassifier.create_from_options(options)

    # STEP 3: Classify the input text.
    classification_result = classifier.classify(INPUT_TEXT)

    # STEP 4: Process the classification result. In this case, print out the most likely category.
    top_category = classification_result.classifications[0].categories[0]
    return top_category.score
  except:
    return 0
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file into a DataFrame
daily_schedule = pd.read_excel('/content/drive/MyDrive/daily_schedule.xlsx')

# Calculate the counts of each status
status_counts = daily_schedule['Status'].value_counts()

# Calculate percentages
total_activities = status_counts.sum()
percentage_completed_on_time = (status_counts.get('On Time', 0) / total_activities) * 100
percentage_not_completed_on_time = (status_counts.get('Delayed', 0) / total_activities) * 100
percentage_canceled = (status_counts.get('Cancelled', 0) / total_activities) * 100
percentage_missed = (status_counts.get('Missed', 0) / total_activities) * 100
percentage_not_recorded = (status_counts.get('Not Recorded', 0) / total_activities) * 100

# Print results
print("Percentage of activities completed on time:", percentage_completed_on_time)
print("Percentage of activities not completed on time:", percentage_not_completed_on_time)
print("Percentage of activities canceled:", percentage_canceled)
print("Percentage of activities Missed:", percentage_missed)
print("Percentage of activities Not Recorded:", percentage_not_recorded)

# Create a pie chart
labels = ['Completed on time', 'Delayed', 'Canceled', 'Missed', 'Not Recorded']
sizes = [percentage_completed_on_time, percentage_not_completed_on_time, percentage_canceled, percentage_missed, percentage_not_recorded]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Activity Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
